## 실습-데이터클리닝
1. 의미 없는 데이터를 제거하기
1. 유전자 발현이 있는 데이터만 남기기(모두 0인 값 제거)
1. Treat 1-4, 2-5, 3-6은 비슷한 데이터이므로 평균값만 남기기
1. 결측치가 있으면 해당 row는 제거함

의미 없는 데이터: 인덱스, unnamed:0, 모두 0인 값, 해당 row

```python
df=df.drop('Unnamed: 0', axis=1) #drpo-지우는 함수, 지울이름, 축(axis)) 
```

```python
df
```

```python
for i in df.index:
    df.loc[i]
```

```python
df.iloc[i,1:]
```

```python
df.iloc[15,1:].sum()==0
```

```python
mask = []

for i in df.index:
    if df.iloc[i,1:].sum() == 0:
        mask.append(True)
    else:
        mask.append(False)
```

```python
df = df[[not m for m in mask]]
```

```python
df
```

```python
df[['Treat1','Treat4']]
```

```python
df[['Treat1','Treat4']].mean(axis=1) #평균
```

```python
df['Treat1_4_mean']=df[['Treat1','Treat4']].mean(axis=1)
```
