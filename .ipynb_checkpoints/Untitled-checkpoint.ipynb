## 실습: while문, for문 상호 변환
* 아래 코드를 다른 형태의, 루프문으로 변경하자

```python
x='99,25,4,5,6,7,8'
ind=0

while 1:
    if ind==0 or ind==1:
        print(x[ind][-1])
    
    else:
        print(x[ind])

        
        
    ind+=1
        
    if ind==7:
        break
```

```python
x='99,25,4,5,6,7,8'

for i in range(0,7):
    print(x[i])
```

```python
x='99,25,4,5,6,7,8'
for i in range(0,7):
    if i==0 or i==1:
        print(x[i][-1])
    else:
        print(x[i])
```

## 이중루프
 for문, while문은 모두 이중루프 이상이 가능하다

```python
list_ex=[]

for i in range(10):
    for j in range(10):
        list_ex.append([1,2,3])
```

```python
list_ex
```

```python
list_ex=[]

for i in range(10):
    list_ex.append([1])
    
    for  in range(10):
        list_ex.append([2,3])
```

## 실습
이중루프 만들기
루프문 5개를 이용한 이중루프를 만들자
단 루프문 3개는 서로 관련이 있어야 한다(같은 변수를 공유하거나 해당 변수의 값을 받아쓴다거나)
어떤 루프문 2개는 서로 관련이 없어야 한다(변수 등을 공유하지 않고, 서로의 값을 영향을 끼치지 못함)
하나 이상의 루프문은 다른 루프문 하나 이상을 감싼 형태여야 한다
두 덩어리의 이중루프가 있어도 무방하다
예제의 종류 및 연산방법은 수강생이 정한다

```python
name = '민수'
print(f'나의 이름은 {name}입니다.')
```

```python
for i in range(1,11):
    for j in range(1,11):
        print(i*j, end = '\t')
    print('')
```
