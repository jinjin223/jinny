## 실습-통계분석
* 위 그래프에 해당하는 데이터의 T-test 진행하기
* 단 그룹 선택은 임의의 한 쌍으로 진행

```python
from scipy import stats
```

```python
df_1027=df_mask['업체아이디']==1027
```

```python
stats.shapiro(df_1027) #pvlue거의 0에 가까움, 정규분포 아니라고 봄
```

이 검정의 귀무가설음 모집단이 정규분포를 따른다는 것, p값이 선택한 알파 수준보다 작으면 귀무가설이 기각되고 테스트된 데이터가 정규 분포를 따르지 않는다는 증거가 있습니다. 반면에 p값이 선택한알파 수준보다 크면 귀무 가설을 기각할 수 없습니다.

대부분의 통계적 유의성 테스트와 마찬가지로 샘플 크기가 충분히 크면 이 테스트는 귀무가설에서 아주 사소한 일탈도 감지할 수 있습니다. 즉 통계적으로 유의미한 효과가 있을 수 있지만, 너무 작아서 실질적인 의미가 없을 수있다. 따라서 효과 크기에 대한 추가조사(예: 이 경우 Q-Q 플롯)

```python
stats.probplot(df_1027, plot=plt)
plt.show()
```

```python
df_1040=df_mask['업체아이디']==1040
```

```python
stats.probplot(df_1040, plot=plt)
plt.show()
```

```python
stats.ttest_ind(df_1027,df_1040)
```
